This project focuses on analyzing a dataset of Canadian immigrants over a range of years. Using the scripting layer to create a stacked area plot of the 5 countries that contributed the least to immigration to Canada **from** 1980 to 2013. Use a transparency value of 0.45.


** Here the dtypes of the dataframe data_can has the dtype -> object so i will change the given dtype of the indeces -> years to 'int'

```python
data_can.tail()
# Storing the value of the least total of the dataframe to a variable 

data_can_least5 = data_can.tail()

# Transposing the dataframe so that we get the years as the index values 

data_can_least5 = data_can_least5[years].transpose()
# data_can_least5.index

# Mapping the given index of years to int as it is stored as str-> 'O'

data_can_least5.index = data_can_least5.index.map(int)

# Adding some of the properties for the attributes of the given data
colors = ['#03dbfc','#ff0a91','#fb0aff','#9d0aff','#50ff0a']
data_can_least5.plot(kind='area',
                     stacked=False,
                     alpha=0.5,
                     figsize=(10,9),
                     color=colors,
                     linewidth=2.5)
# Setting of the various list of the internal properties of the given plotting 


plt.title("Mapping of Least Immigrants VS Countries")
plt.title("IMMIGRANTS DATA")
plt.xlabel("Years of the immigration",
           color='r')
plt.ylabel("No of Immigrants in that year")
plt.show()
plt.savefig("Area Plot for Immigrants Data.png")
```

```python
# Plotting the histogram of the given data to know the total frequency of the immigrants population at year '2013'
data_2013 = data_can['2013']

# Setting up the xticks_bins for the categorization of the distribution of the data 

count , bin_edges = np.histogram(data_2013)
# Setting up the figuresize
plt.figure(figsize=(10,10))

plt.hist(data_2013,
         bins=len(bin_edges) - 1,
         color='b',
         label='Frequency Distribution - 2013'
         )

# Labeling of the graph 
plt.title("Histogram to check the frequency distribution",
        fontdict={
        'family': 'serif',
        'color': 'black', 
        'fontsize': 18
        })

plt.xticks(bin_edges, rotation=45)

plt.xlabel("No of immigrants for year 2013",
           color='red')
plt.ylabel("Frequency of Countries",
           color='red')
plt.legend(
    labelcolor='blue'
)
plt.show()
plt.savefig("Histogram for frequency check.png")
```

**Let's start off by analyzing the effect of Iceland's Financial Crisis:**

The 2008 - 2011 Icelandic Financial Crisis was a major economic and political event in Iceland. Relative to the size of its economy, Iceland's systemic banking collapse was the largest experienced by any country in economic history. The crisis led to a severe economic depression in 2008 - 2011 and significant political unrest.

## Compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.

```python
# Fetch the data for the label named ' Iceland ' from the dataframe

data_iceland = data_can.loc['Iceland',years]
data_iceland
# Here the year along with the frequency of the immigrants of Iceland over the years are drawn
# 1980 -> 17 means that in 1980, over 17 immigrants migrated to Canada ( maybe due to the financial conditions or so..)

data_iceland.plot(kind='bar',
                  figsize=(10,7),
                  label='Data Iceland'
                  )
plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Icelandic immigrants to Canada from 1980 to 2013') # add title to the plot
plt.legend()
plt.savefig("BarPlots_of_Immigrants.png")
```
